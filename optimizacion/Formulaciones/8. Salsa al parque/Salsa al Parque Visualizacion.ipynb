### Notación:

#### Conjuntos

* $A$: artistas
* $H$: horas

#### Parámetros

* $d_{i}$: duración de la presentación del artista $i\in A$
* $a_{it}$: audiencia del artista $i\in A$ durante la hora $t\in H$

#### Variables

* $x_{it}:\begin{cases}1\text{,}&\text{ si el artista }i\in A\text{ inicia su presentación en la hora }t\in H \\ 0\text{,}& \text{ d.l.c.}  \end{cases}$
* $y_{it}:\begin{cases}1\text{,}&\text{ si el artista }i\in A\text{ se presenta durante la hora }t\in H \\ 0\text{,}& \text{ d.l.c.}  \end{cases}$


### Formulación

**a.** Escriba la(s) restricción(es) lineal(es) que garantiza(n) que la cantidad de artistas presentandose en simultaneo a cada hora $h \in H$ sea menor o igual a $k$.

$$
\begin{align*}
    \sum_{i \in A}y_{i,t} & \leq k, \forall t \in H
\end{align*}
$$

**b.** Escriba la(s) restricción(es) lineal(es) que garantiza(n) que cada artista $i \in A$ inicia su presentación una vez.

$$
\begin{align*}
    \sum_{t \in H}x_{i,t} & = 1, \forall i \in A
\end{align*}
$$

**c.** Escriba la(s) restricción(es) lineal(es) que garantiza(n) que cada artista $i \in A$ se presenta en horas consecutivas y que no presente durante horas previas a la hora de inicio.

- El artista $i \in A$ se presenta en horas consecutivas desde el inicio de su presentación:

$$
\begin{align*}
    \sum_{j = t}^{t+d_{i}-1}y_{i,j} & \geq d_{i}x_{i,t}, \forall i \in A, t \in H \ |\  t + d(i) - 1 \leq |H|
\end{align*}
$$

- El artista $i \in A$ se presenta exactamente las horas que dura su presentación:

$$
\begin{align*}
    \sum_{t \in H}y_{i,t} & = d_{i}, \forall i \in A
\end{align*}
$$

- El artista $i \in A$ no puede iniciar su presentación si el tiempo restante es insuficiente:

$$
\begin{align*}
    x_{i,t} & = 0, \forall i \in A, t \in H \ | \ t + d_{i} - 1 > |H|
\end{align*}
$$

**d.** Escriba la(s) restricción(es) que describe(n) la naturaleza de la(s) variable(s) incluida(s) en el modelo.

$$
\begin{align*}
    x_{it}, y_{it} & \in \{0,1\}, \forall i \in A, t \in H
\end{align*}
$$

**e.** Escriba la función objetivo.

$$
\begin{align*}
    \max \sum_{i \in A} \sum_{t \in H} a_{it}y_{it}
\end{align*}
$$

### Implementación
**f.** Resuelva el modelo planteado utilizando la librería de PulP en Python. ¿Cuál es la solución
óptima del problema? 

```python
#Se importa la libreria PulP
import pulp as lp
#Se importa la librería de gráficos Matplotlib
import matplotlib.pyplot as plt

#-----------------
# Conjuntos
#-----------------
#Artistas
A = [str(i + 1) for i in range(5)]

#Horas
H = [t + 1 for t in range(12)]

#-----------------
# Parámetros
#-----------------
#Se cargan todos los parámetros del modelo
d = {# Artista: duración de la presentación
    "1": 2,
    "2": 3,
    "3": 2,
    "4": 3,
    "5": 2}

a = {# (Artista, Hora): tamaño de la audiencia
    ('1',1): 1197, ('1',2): 2578, ('1',3): 1350, ('1',4): 2729, ('1',5): 2679, ('1',6): 2946, ('1',7): 1563, ('1',8): 2998, ('1',9): 2879, ('1',10): 1126, ('1',11): 2559, ('1',12): 1129,
    ('2',1): 1034, ('2',2): 2009, ('2',3): 1515, ('2',4): 2254, ('2',5): 1624, ('2',6): 1277, ('2',7): 2010, ('2',8): 2563, ('2',9): 1384, ('2',10): 2695, ('2',11): 2053, ('2',12): 2585,
    ('3',1): 1058, ('3',2): 1842, ('3',3): 1443, ('3',4): 1995, ('3',5): 2819, ('3',6): 1998, ('3',7): 1401, ('3',8): 1708, ('3',9): 1346, ('3',10): 1674, ('3',11): 1816, ('3',12): 1032,
    ('4',1): 1082, ('4',2): 1206, ('4',3): 2366, ('4',4): 1013, ('4',5): 2762, ('4',6): 1479, ('4',7): 2496, ('4',8): 1509, ('4',9): 1673, ('4',10): 1350, ('4',11): 1128, ('4',12): 2669,
    ('5',1): 2801, ('5',2): 1003, ('5',3): 2965, ('5',4): 2295, ('5',5): 1147, ('5',6): 2156, ('5',7): 1058, ('5',8): 2857, ('5',9): 1968, ('5',10): 2764, ('5',11): 2153, ('5',12): 1518}

k = 2 #máxima cantidad de artistas presentando en simultaneo

#-------------------------------------
# Creación del objeto problema en PuLP
#-------------------------------------
#Crea el problema para cargarlo con la instancia 
problema = lp.LpProblem("Scheduling-Salsa-al-Parque",lp.const.LpMaximize)

#-----------------------------
# Variables de Decisión
#-----------------------------
x = lp.LpVariable.dicts('x',[(i, t) for i in A for t in H],lowBound = 0,cat = lp.const.LpBinary) #si el artista a en A inicia su presentación en la hora h in H
y = lp.LpVariable.dicts('y',[(i, t) for i in A for t in H],lowBound = 0,cat = lp.const.LpBinary) #si el artista a en A se presenta durante la hora h in H

#-----------------------------
# Función objetivo
#-----------------------------
#Crea la expresión de maximización de audiencias
problema += lp.lpSum(a[(i, t)]*y[i, t] for i in A for t in H), "Audiencia_total"

#-----------------------------
# Restricciones
#-----------------------------
for t in H:
    #la cantidad de artistas presentandose en simultaneo a cada hora h en H es menor o igual a k.
    problema += lp.lpSum(y[(i, t)] for i in A) <= k
    
for i in A:
    #cada artista i en A inicia su presentación una vez.
    problema += lp.lpSum(x[(i, t)] for t in H) == 1
    
for i in A:
    for t in H:
        if t + d[i] - 1 <= len(H):
            #cada artista i en A se presenta en horas consecutivas y no presenta durante horas previas a la hora de inicio.
            problema += lp.lpSum(y[(i, j)] for j in range(t, t + d[i])) >= d[i] * x[(i, t)]
            
for i in A:
    #cada artista i en A debe presentarse por exactamente la cantidad de horas que le corresponden
    problema += lp.lpSum(y[(i,t)] for t in H) == d[i]
        
for i in A:
    for t in H:
        #cada artista no puede iniciar su presentación si el tiempo es insuficiente para terminar
        if t + d[i] - 1 > len(H):
            problema += x[(i,t)] == 0
        
#-----------------------------
# Invocar el optimizador
#-----------------------------
#Optimizar el modelo con CBC (default de PuLP)
problema.solve()

#-----------------------------
#    Imprimir resultados
#-----------------------------
#Imprimir estado final del optimizador
print(f"Estado (optimizador): {lp.LpStatus[problema.status]}")

#Audiencia total alcanzada
print(f"Audiencia Total: {lp.value(problema.objective)*1000}")
print()

#Imprimir variables de decisión
print("Variables de decisión")
print("", end = "\t")
for t in H:
    print(t, end = "\t")
print()

for i in A:
    print(i, end = "\t")
    for t in H:
        print(int(round(y[(i, t)].value(), 0)), end = "\t")
    print()
    
#-----------------------------
#    Graficar resultados
#-----------------------------
#Guardamos en listas las horas de inicio y fin de cada artista
comienzo = [sum(t * x[i, t].value() for t in H) for i in A]
fin = [comienzo[i] + d[str(i + 1)] for i in range(len(A))]

#Graficamos la solución por medio de la referencia plt
plt.hlines(A, comienzo, fin, linewidth = 35)
plt.xlabel(xlabel = "Hora")
plt.ylabel(ylabel = "Artista")
plt.title('Itinerario de Presentaciones')
plt.xlim([0.9, 13.1])
plt.xticks(H + [13])
plt.margins(0.1)
plt.show()
```
